```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from xgboost import XGBRegressor
```

```python
from sklearn.linear_model import Ridge

from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import LeaveOneGroupOut
```

```python
DF1 = pd.read_csv('train.csv')

DF2 = pd.read_csv('test.csv')
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(DF2)
data2 = pd.DataFrame(x_scaled)
```

```python
data2 = DF2.copy()
```

```python
data1 = DF1.copy()
```

```python
data1
```

```python
y = DF1['target']
```

```python
x = data1.drop('target',axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.70, random_state=314)
```

```python
model1v = XGBRegressor(max_depth=5,  n_estimators=1000, learning_rate=0.001, subsample=0.7, alpha=0.5, random_state=314)                           
model1v.fit(x_train, y_train, verbose=0)
oof_pred1 = model1v.predict(x_train)
oof_pred1 = np.clip(oof_pred1, 0.30, y.max())
oof_pred1
```

```python
model1v.feature_importances_
```

```python
axis_x  = x.columns.values
axis_y = model1v.feature_importances_

plt.style.use('seaborn-whitegrid') 
plt.figure(figsize=(20, 8), facecolor='lightgray')

plt.scatter(axis_x, axis_y, s=120, label='target') 
plt.legend(fontsize=12, loc=2)
plt.show()
```

```python
pred1 = model1v.predict(data2)
pred1 = np.clip(pred1, 0.30, y.max())
display(pred1, pred1.shape)
sample = pd.read_csv('sample_submission.csv')
sub_xgb = sample.copy()

sub_xgb['target'] = pred1
```
